```python
import matplotlib.pyplot as plt
import numpy as np
import os
import json
```

```python
baseline = {'MSE':[4.47, 4.20, 5.10, 6.09, 9.12],
            'FID': [38.87, 38.14, 47.08, 62.48, 77.51],
            'SFID': [220.82, 209.21, 232.46, 259.21, 234.52]}

euclidian = {'MSE': [4.12, 3.33, 3.61, 3.72, 5.94],
             'FID': [37.64, 33.71, 34.03, 37.25, 48.36],
             'SFID': [216.57, 201.55, 206.33, 244.66, 222.77]}


cosine = {'MSE': [3.98, 3.31, 4.95, 5.20, 7.28],
          'FID': [37.14, 32.80, 41.24, 42.25, 50.33],
          'SFID': [218.57, 213.27, 201.92, 217.63, 224.17]}


gumble = {'MSE': [4.12, 3.77, 4.18, 6.71, 7.27],
          'FID': [37.23, 33.20, 42.12, 58.25, 65.12],
          'SFID': [199.20, 183.08, 201.14, 227.18, 222.12]}
```

```python
# set width of bar
barWidth = 0.25
fig = plt.subplots(figsize =(12, 8))


# Set position of bar on X axis
br1 = 2*np.arange(len(baseline['MSE']))
br2 = [x + barWidth for x in br1]
br3 = [x + barWidth for x in br2]
br4 = [x + barWidth for x in br3]


colors = plt.cm.BuPu(np.linspace(0, 0.5, 4))

# Make the plot
plt.bar(br1, baseline['MSE'], color = colors[0], width = barWidth,
        edgecolor ='grey', label ='SA')
plt.bar(br2, euclidian['MSE'], color = colors[1], width = barWidth,
        edgecolor ='grey', label ='BSA-Euclidian')
plt.bar(br3, cosine['MSE'], color = colors[2], width = barWidth,
        edgecolor ='grey', label ='BSA-Cosine')
plt.bar(br4, gumble['MSE'], color = colors[3], width = barWidth,
        edgecolor ='grey', label ='BSA-Gumble')



# Adding Xticks
plt.xlabel('Iterations', fontweight ='bold', fontsize = 15)
plt.ylabel('MSE', fontweight ='bold', fontsize = 15)
plt.xticks([2*r + 1.5*barWidth for r in range(len(baseline['MSE']))],
        ['1', '2', '3', '4', '5'])
 
plt.legend()
plt.savefig('mse-lemma1.png')
plt.show()
```
